#### Data imputation of 0's in every feature

```python
df['Pregnancies'] = df['Pregnancies'].replace(0,df['Pregnancies'].median())
df['Glucose'] = df['Glucose'].replace(0,df['Glucose'].mean())
df['BloodPressure'] = df['BloodPressure'].replace(0,df['BloodPressure'].mean())
df['SkinThickness'] = df['SkinThickness'].replace(0,df['SkinThickness'].median())
df['Insulin'] = df['Insulin'].replace(0,df['Insulin'].median())
df['DiabetesPedigreeFunction'] = df['DiabetesPedigreeFunction'].replace(0,df['DiabetesPedigreeFunction'].median())
df['BMI'] = df['BMI'].replace(0,df['BMI'].mean())
df['Age'] = df['Age'].replace(0,df['Age'].median())
```

```python
df.head(20)
```

#### x = input feature
#### y = target value


```python
x = df.drop(columns='Outcome',axis=1)
y = df['Outcome']
```

#### Outlier Detection

```python
fig,ax = plt.subplots(figsize=(15,15))
sns.boxplot(data=x,ax=ax)
plt.savefig('boxplot.jpg')
```

```python
print(x.shape)
print(y.shape)
```

```python
cols = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age']
for col in cols:
    Q1 = x[col].quantile(0.25)
    Q3 = x[col].quantile(0.75)
    IQR = Q3-Q1
    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR
    mask = (x[col]>=lower) & (x[col]<=upper)
```

```python
x_outlier = x[mask]
y_outlier = y[mask]
```

```python
print(x_outlier.shape) 
print(y_outlier.shape)
```

### Standardization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scale = scaler.fit_transform(x_outlier)
```

```python
plt.figure(figsize=(15,15))
sns.boxplot(data=x_scale)
```

### Model Training : Splitting of data

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scale,y_outlier,test_size=0.33,random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
y_train.value_counts()
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
x_train_resample,y_train_resample = smote.fit_resample(x_train,y_train)
print('\nResampled class distribution')
print(pd.Series(y_train_resample).value_counts())
```

#### Implementation of Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
classification = LogisticRegression()
classification.fit(x_train_resample,y_train_resample)
```

### Model Prediction 

```python
pred_value = classification.predict(x_test)
print(pred_value)
```

### Model Evaluation

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,pred_value)
```

```python
from sklearn.metrics import classification_report
target = ['Non Diabetic','Diabetic']
print(classification_report(y_test,pred_value,target_names=target))
```
